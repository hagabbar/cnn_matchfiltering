```python
import numpy as np
import cPickle as pickle

import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# choose run
snr = 8
run = 14
# load history
history_path = './history/SNR{0}/run{1}'.format(snr,run)
history = np.load('{0}/history.npy'.format(history_path))
```

```python
# retrieve relevant data
loss = history.item().get('loss')
acc = history.item().get('acc')
val_loss = history.item().get('val_loss')
val_acc = history.item().get('val_acc')
lr = history.item().get('lr')
epochs = np.arange(0,len(loss),1)
```

```python
colours = ['indigo', 'c', 'darkorange']
# plot everything
fig1 , axs = plt.subplots(3,1, sharex = 'col')
axs = axs.ravel()
#plot loss
axs[0].plot(epochs,loss, c=colours[0])
axs[0].plot(epochs, val_loss, c=colours[1])
axs[0].set_ylabel('Loss')
# plot accuracy
axs[1].plot(epochs, acc, c=colours[0])
axs[1].plot(epochs, val_acc, c=colours[1])
axs[1].set_ylabel('Accuracy')
# plot learning rate
axs[2].plot(epochs, lr, c=colours[0])
axs[2].set_xticklabels(epochs)
axs[2].set_xlabel('Epochs')
axs[2].set_ylabel('Learning rate')
plt.show()
fig1.savefig('loss.pdf', format='pdf')
```
